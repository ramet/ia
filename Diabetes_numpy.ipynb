```python
import matplotlib.pyplot as plt
import numpy as np

data = np.genfromtxt('diabetes.csv', dtype=float, delimiter=',', names=True)

data
```

```python
data.shape
```

```python
# Moyenne
np.mean(data['Pregnancies'])
```

```python
# Ecart type
np.std(data['Pregnancies'])
```

```python
plt.grid()
plt.hist(data['Pregnancies'])
```

```python
less_than_40 = data[data['Age'] < 40]
less_than_40
```

```python
plt.scatter(less_than_40['Age'], less_than_40['Pregnancies'])
plt.xlabel('Age')
plt.ylabel('Pregnancies')
```

```python
less_than_40_0 = less_than_40[less_than_40['Outcome'] == False]
less_than_40_1 = less_than_40[less_than_40['Outcome'] == True]

plt.scatter(less_than_40_0['Age'], less_than_40_0['Pregnancies'], c='green', label='Outcome=0')
plt.scatter(less_than_40_1['Age'], less_than_40_1['Pregnancies'], c='red', label='Outcome=1')

plt.xlabel('Age')
plt.ylabel('Pregnancies')
plt.legend()
plt.show()
```
